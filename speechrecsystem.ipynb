```python
!pip install SpeechRecognition
!apt-get install -y ffmpeg

import wave
import speech_recognition as sr
from pydub import AudioSegment
from google.colab import files

def record_audio(filename='recorded_audio.wav', duration=5):
    audio = AudioSegment.silent(duration * 1000)

    print("Recording... Speak for a few seconds.")
    audio.export(filename, format="wav")

def recognize_speech(audio_path):
    recognizer = sr.Recognizer()

    with sr.AudioFile(audio_path) as source:
        audio_data = recognizer.record(source)

    try:
        print("Recognizing...")
        text = recognizer.recognize_google(audio_data)
        print("You said:", text)
    except sr.UnknownValueError:
        print("Sorry, could not understand audio.")
    except sr.RequestError as e:
        print(f"Error with the speech recognition service; {e}")

# Record audio
record_audio()

# Upload the 'recorded_audio.wav' file to Colab
print("Now, upload the 'recorded_audio.wav' file.")
uploaded = files.upload()

# Perform speech recognition on the uploaded file
if 'recorded_audio.wav' in uploaded:
    recognize_speech('recorded_audio.wav')
else:
    print("No file uploaded.")
```
